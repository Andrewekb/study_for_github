```python
import pandas as pd
```

Здесь попытался создать X_train, y_train на всей выборке, не удаляя пропуски в Age

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['Embarked'])
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.Embarked = le.fit_transform(df['Embarked'])
```

```python
df.Sex = le.fit_transform(df['Sex'])
```

```python
df.head()
```

```python
df = df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

Формирую выборку для обучения - 70%, без пропусков

```python
train = df.dropna(subset=['Age']).iloc[0:int((df.shape[0]*0.7))]
```

```python
X_train = train.drop(['Age'], axis=1)
```

```python
X_train
```

```python
y_train = train.Age
```

```python
y_train
```

Тестовая выборка - вычитаю из общего df 70 процентов отобранных для обучения

```python
test = df[~df.index.isin(X_train.index)]
```

```python
X_test = test.drop(['Age'], axis=1)
```

```python
X_test
```

```python
y_test = test.Age
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print('Ошибка на обучающей выборке', mean_absolute_error(y_train, lr.predict(X_train)))
```
